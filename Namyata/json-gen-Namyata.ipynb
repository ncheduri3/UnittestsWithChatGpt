```python
import pandas as pd
import os
import sys
```

```python
len(os.listdir())
```

```python
files = []
for file in os.listdir():
    if os.path.isfile(file):
        if file[-3:] == '.py' and file != 'Namyata_concatenated.py':
            files.append(file)
```

```python
def sortkey(s):
    return s.lower()
files = sorted(files, key=sortkey)
len(files)
```

```python
function_names = [] # DONE
headers = [] # DONE
docstrs = [] # DONE
function_code = [] # DONE
test_code = [] # DONE

for i, file in enumerate(files):
    with open(file, 'r') as filein:
        lines = filein.readlines()
        commentlines = []
        j=0
        for j, line in enumerate(lines):
            if line[0] == '#':
                found = True
                break
            if line[:3] == 'def':
                found=False
                break
        if found:
            for k, line in enumerate(lines[j:]):
                if line[0] != '#':
                    break
                commentlines.append(line)
            docstrs.append(''.join([line[1:] for line in commentlines]))

        else:
            docstrs.append('')

        for m, line in enumerate(lines):
            if line[:3] == 'def':
                j = m
                break
        
        headers.append(lines[j])
        function_names.append(lines[j].split()[1].split('(')[0])

        codelines = []
        for m, line in enumerate(lines[j+1:]):
            if line[:5] == 'class':
                j += m + 1
                break
            codelines.append(line)
        
        codestripped = codelines
        for k, line in enumerate(codelines[::-1]):
            if line != '\n':
                break
            else:
                codestripped = codestripped[:-1]

        function_code.append(''.join(codestripped))

        testlines = []
        for m, line in enumerate(lines[j+1:]):
            if line[:26] == 'if __name__ == \'__main__\':' or line[:26] == 'if __name__ == \"__main__\":':
                j += m + 1
                break
            testlines.append(line)
        
        teststripped = testlines
        for k, line in enumerate(testlines[::-1]):
            if line != '\n':
                break
            else:
                teststripped = teststripped[:-1]
        if teststripped[0][0] == '\n':
            teststripped[0] = teststripped[0][1:]
        test_code.append(''.join(teststripped))
        
```

```python
df = pd.DataFrame({'function_name': function_names,
                   'header': headers,
                   'docstr': docstrs,
                   'code': function_code,
                   'test': test_code
                   })
df
```
