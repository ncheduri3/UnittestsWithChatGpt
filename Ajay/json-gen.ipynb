```python
import pandas as pd
import os
import sys
```

```python
dirs = []
for dir in os.listdir():
    if os.path.isdir(dir):
        dirs.append(dir)
len(dirs)
    
```

```python
def sortkey(s):
    if s[1] == '_':
        return int(s[0])
    elif s[2] == '_':
        return int(s[:2])
    
dirs = sorted(dirs, key=sortkey)
```

```python
functions = []
tests = []

for dir in dirs:
    for file in os.listdir(dir):
        if file[-3:] == '.py':
            if file[:5] == 'test_':
                tests.append(file)
            else:
                functions.append(file)
len(tests), len(functions), tests[3], functions[3]
```

```python
function_names = []
headers = []
docstrs = [''] * len(dirs)
function_code = []
test_code = []

for i, function, test, dir in zip(range(len(dirs)), functions, tests, dirs):
    with open(f'{dir}/{function}', 'r') as function_f:
        lines = function_f.readlines()
        headerlines = []
        j = 0
        while lines[j][:3] != 'def':
            headerlines.append(lines[j])
            j+=1
        headerlines.append(lines[j])
        if len(headerlines) == 1: headerlines = [f'\n{headerlines[0]}']
        headers.append(''.join(headerlines))
        function_names.append(lines[j].split()[1].split('(')[0])
        function_code.append(''.join(lines[j+1:]))
        
    with open(f'{dir}/{test}', 'r') as test_f:
        lines = test_f.readlines()
        j = 0
        while lines[j][:5] != 'class':
            j+=1
        k = len(lines) - 1
        while lines[k][:26] == 'if __name__ == \'__main__\':' or lines[k][:26] == 'if __name__ == \"__main__\":':
            k -= 1
        test_code.append(''.join(lines[j+1:k - 2]))
```

```python
import json

func_json = json.dumps(function_names)
with open('func_json.json', 'w') as fileout:
    fileout.write(func_json)
```

```python
df = pd.DataFrame({'function_name': function_names,
                   'header': headers,
                   'docstr': docstrs,
                   'code': function_code,
                   'test': test_code
                   })
df
```

```python
df.to_json('unit_tests.json', orient='records', lines=True)
```
